```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
from sklearn.decomposition import PCA

%pylab inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston().data
```

```python
boston.keys()
```

```python
 print "Shape of Boston Dataset : ", boston.data.shape
print "\ndatatypes of dataset:\n",boston.data.dtype
```

```python
bos_df=pd.DataFrame(boston,columns=load_boston().feature_names)
bos_df.head()
```

```python
pd.scatter_matrix(bos_df, alpha=0.8, figsize=(14, 14));
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(bos_df)
bos_standard=sc.transform(bos_df)
```

```python
bos_df.columns.size
```

```python
#bos_standard
#performing PCA
for i in range(bos_df.columns.size):
    pca_bos=PCA(n_components=i)
    pca_bos.fit(bos_standard)
    print pca_bos.explained_variance_ratio_.sum()
```
